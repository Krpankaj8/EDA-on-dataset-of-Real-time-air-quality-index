## Exploratory Data Analysis of Air Quality Across Indian Cities

#### This project aims to analyze air quality data collected from various Indian cities. 
#### It explores pollutant trends, identifies pollution hotspots, and provides visual summaries to help 
#### understand the extent of air pollution across regions. 
#### The analysis uses multiple objectives to extract meaningful insights from the dataset.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DATASET INT 375.csv')
```

```python
print(df.head(5))
```

### Handling the missing values

```python
for col in ['pollutant_min', 'pollutant_max', 'pollutant_avg']:
    mean_val = df[col].mean()
    df[col] = df[col].fillna(mean_val)
```

```python
print(df.head)
```

#### Renaming the column longitude with pollutant

```python
df.rename(columns={'longitude.1': 'pollutant'}, inplace=True)
```

```python
print(df.head(5))
```

#### Checking the missing values

```python
print("NA values in dataset:", df.isnull().values.any())
```

### Objective 1

#### Analysis of Air quality across Indian cities

#### Identified the top 20 most polluted cities in India based on average pollutant concentrations, and visualized them using a bar chart for comparative analysis.

```python
city_avg_pollution = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
city_avg_pollution.head(20).plot(kind='bar', color='#C54E48')
plt.title('Top 20 Most Polluted Cities in India (by Avg Pollutant Level)')
plt.ylabel('Average Pollutant Level')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

### Objective 2

#### Correlation Analysis Between Pollutants

#### A heatmap was created to analyze the relationships between different pollutants and understand how they vary together.

```python

pivot_df = df.pivot_table(values='pollutant_avg', index=['state', 'city', 'station'], columns='pollutant')

# Generate heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Different Pollutants')
plt.show()
```

### Objective 3

#### Cities/States with Highest & Lowest Pollution

#### Identified the most and least polluted cities and states in India based on average pollutant levels.

```python
state_pollution = df.groupby('state')['pollutant_avg'].mean().sort_values()
city_pollution = df.groupby('city')['pollutant_avg'].mean().sort_values()

print("Top 5 Least Polluted States:")
print(state_pollution.head())
print("\nTop 5 Most Polluted States:")
print(state_pollution.tail())

print("\nTop 5 Least Polluted Cities:")
print(city_pollution.head())
print("\nTop 5 Most Polluted Cities:")
print(city_pollution.tail())
```

### Objective 4

#### Most Common Pollutants

#### Analyzed which pollutants are monitored most frequently across the dataset using bar charts.

```python
pollutant_counts = df['pollutant'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=pollutant_counts.values, y=pollutant_counts.index, palette='viridis')
plt.title('Most Frequently Monitored Pollutants')
plt.xlabel('Number of Records')
plt.ylabel('Pollutant')
plt.tight_layout()
plt.show()
```

### Objective 5

#### Number of Monitoring Stations Per State

#### Visualized the number of air quality monitoring stations in each state to assess data coverage.

```python
stations_per_state = df.groupby('state')['station'].nunique().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
stations_per_state.plot(kind='bar', color='teal')
plt.title('Number of Monitoring Stations in Each State')
plt.ylabel('Number of Stations')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

### Objective 6

#### Most Polluted Monitoring Stations

#### Highlighted the top 10 monitoring stations with the highest average pollutant concentrations.

```python
station_pollution = df.groupby('station')['pollutant_avg'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
station_pollution.head(10).plot(kind='bar', color='salmon')
plt.title('Top 10 Most Polluted Monitoring Stations')
plt.ylabel('Average Pollutant Level')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

### Objective 7

#### Air Quality Summary for Each State


#### Provided a quick summary of mean, minimum, and maximum pollution levels for each state.

```python
state_summary = df.groupby('state')['pollutant_avg'].agg(['mean', 'min', 'max']).sort_values(by='mean', ascending=False)

print("Air Quality Summary by State (mean, min, max):")
print(state_summary.head(10))  # Top 10 for readability
```

### Objective 8

#### Distribution of Pollutant Values

#### Used a box plot to display the spread and variability of pollutant levels for different pollutant types.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pollutant', y='pollutant_avg', data=df, palette='Set2')
plt.title('Distribution of Average Pollutant Levels by Type')
plt.ylabel('Pollutant Average')
plt.xlabel('Pollutant')
plt.tight_layout()
plt.show()
```

### Objective 9

#### Compare Pollution in Metro Cities

#### Compared average pollutant levels among major metro cities like Delhi, Mumbai, and Bengaluru to spot urban pollution trends.

```python
metro_cities = ['Delhi', 'Mumbai', 'Bengaluru', 'Chennai', 'Kolkata', 'Hyderabad']
metro_df = df[df['city'].isin(metro_cities)]

plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='pollutant_avg', data=metro_df, estimator='mean', palette='coolwarm')
plt.title('Average Pollutant Level in Major Indian Metro Cities')
plt.ylabel('Average Pollutant Level')
plt.tight_layout()
plt.show()
```
